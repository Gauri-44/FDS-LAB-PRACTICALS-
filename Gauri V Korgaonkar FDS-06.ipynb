```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = sns.load_dataset('mpg').dropna()  # mpg is a regression dataset
data = data.rename(columns={'horsepower': 'feature1', 'weight': 'feature2', 'mpg': 'target'}) # Renaming for experiment
```

```python
data.head()
```

```python
# Step 2: Data Preprocessing
data = data.dropna()

# Split data into features and target variable
X = data[['feature1', 'feature2']]
y = data['target']

```

```python
#Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Step 3: Build and Train the Model
# Initialize linear regression model
model = LinearRegression()
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Step 4: Make Predictions and Evaluate the Model
# Make predictions
y_pred = model.predict(X_test)
```

```python
# Evaluate model performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

```python
# Step 5: Visualize Results
plt.scatter(y_test, y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.show()

```

```python
import pandas as pd

data = pd.read_csv('winequality-red.csv')  # Make sure to adjust the filename if necessary
print(data.head())
```

```python
data.info()
data.describe()
```

```python
data = data.dropna()
```

```python
print(f"Dataset shape: {data.shape}")
```

```python
X = data.drop('quality', axis=1)  # Features: All columns except 'quality'
y = data['quality']  # Target variable: 'quality'
```

```python
# Step 4: Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Step 5:
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Step 6: Make Predictions
y_pred = model.predict(X_test)

# Step 7: Evaluate Model Performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

```python
# Step 8: Visualize Results
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Wine Quality")
plt.ylabel("Predicted Wine Quality")
plt.title("Actual vs Predicted Wine Quality")
plt.plot([y.min(), y.max()], [y.min(), y.max()], '--r')  # Diagonal line
plt.xlim(y.min(), y.max())
plt.ylim(y.min(), y.max())
plt.grid()
plt.show()
```
